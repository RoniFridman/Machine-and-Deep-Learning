```python
def get_infect_prob(G, node, artist):
  node_neigb = list(G.neighbors(node))
  Nt = len(node_neigb)
  Bt = sum([1 for elem in node_neigb if G.nodes[elem]['is_infected'] == 1 ])
  h = G.nodes[node]['music'][artist] 
  if h==0:
    return Bt/Nt
  return (h*Bt)/(1000*Nt)
```

```python
def get_infulencer(G, five_influencer, artist):
  for i in five_influencer:
    G.nodes[i]['is_infected']=1

  infected = five_influencer
  
  for _ in range(5):
    infect_to_add =[]
    for influence in infected:
      influ_neigb = list(G.neighbors(influence))
      for node in influ_neigb:
        rand = random.uniform(0, 1)
        if G.nodes[node]['is_infected'] == 0 and get_infect_prob(G, node, artist)>=rand:
          infect_to_add.append(node)
    for node in infect_to_add:
      G.nodes[node]['is_infected'] = 1
      
    infected = infected.union(set(infect_to_add))
  return len(infected)

```

```python
import pandas as pd
import networkx as nx
import numpy as np

df0 = pd.read_csv('instaglam0.csv')
df_music = pd.read_csv('spotifly.csv')
df_music = df_music.pivot(index='userID', columns=' artistID', values='#plays')
df_music = df_music.replace(np.nan,0).to_dict(orient='index')


G0 = nx.from_pandas_edgelist(df0, source='userID', target='friendID') 
nx.set_node_attributes(G0,0, 'is_infected')
nx.set_node_attributes(G0,df_music, 'music')
```

```python
def create_edges1(graph):
  new_edges = []
  alpha = 0.8
  jc = nx.jaccard_coefficient(graph)
  for u,v in nx.non_edges(graph):
    p = list(nx.jaccard_coefficient(graph,[(u,v)]))[0][2]
    if p==0:
      continue
    else:
      ccpa = alpha*p + (1-alpha)*np.abs(np.random.normal(0,1))
      if ccpa >= 0.58:
        new_edges.append((u,v))
  return new_edges
```

```python
def simulator(graph, nodes, artistID):
  infected_nodes = set(nodes)
  for node in nodes:
    graph.nodes[node]['is_infected'] = 1
  
  for _ in range(6):
    new_nodes = set()
    for infecting in infected_nodes:
      inf_neighbors = graph.neighbors(infecting)
      for node in inf_neighbors:
        if graph.nodes[node]['is_infected']==1:
          continue
        else:
          H = graph.nodes[node]['music'][artistID]
          Nt = graph.neighbors(node)
          Bt = 0
          rand = np.random.rand()
        for neighbor in Nt:
          if graph.nodes[neighbor]['is_infected'] == 1:
            Bt += 1
        if len(list(Nt)) == 0 or Bt == 0 :
               continue
        else:
          if (H==0 and Bt/len(list(Nt) <= rand)) or (H>0 and Bt*H/(len(list(Nt)*1000) <= rand)):
            graph.nodes[node]['is_infected'] = 1
            new_nodes.add(node)
      infected_nodes.union(new_nodes)
      graph.add_edges_from(create_edges(graph))
  
  return infected_nodes
    
    
  
```

```python
def create_edges(graph):
  new_edges = []
  alpha = 0.8
  jc = nx.jaccard_coefficient(graph)
  for u,v,p in jc:
    if p==0 or graph.has_edge(u,v) or graph.has_edge(v,u):
      continue
    else:
      ccpa = alpha*p + (1-alpha)*np.abs(np.random.normal(0,1))
      if ccpa >= 0.58:
        new_edges.append((u,v))
  return new_edges
```
