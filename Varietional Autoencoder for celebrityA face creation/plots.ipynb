```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
cont_df = pd.read_csv('loss_continues.csv', index_col=0)
disc_df = pd.read_csv('loss_discrete.csv', index_col=0)
```

```python
plt.plot(cont_df['Epoch']+1, cont_df['accumulative BCE'], label='Continuous BCE')
plt.plot(disc_df['Epoch']+1, disc_df['accumulative BCE'], label='Discrete BCE')
plt.legend()
plt.title("BCE Losses")
plt.show()
```

```python
plt.plot(cont_df['Epoch']+1, cont_df['accumulative KLD'], label='Continuous KLD')
plt.plot(disc_df['Epoch']+1, disc_df['accumulative KLD'], label='Discrete KLD')
plt.legend()
plt.title("KLD Losses")
plt.show()
```
